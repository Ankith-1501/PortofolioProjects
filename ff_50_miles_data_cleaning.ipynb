```python
import pandas as pd
from geopy.geocoders import Nominatim
from geopy.exc import GeocoderTimedOut
from geopy.extra.rate_limiter import RateLimiter
```

```python
df = pd.read_csv('ff_race_50.csv')
```

```python
df.head()
```

```python
df2 = df.dropna(axis=1).copy()
```

```python
df2.head()
```

```python
df2['Fullname'] = df2['First'] + ' ' + df2['Last']
```

```python
df2.head()
```

```python
df2['Time'] = pd.to_timedelta(df2['Time'])
```

```python
df2['Total_Minutes'] = df2['Time'].dt.total_seconds() / 60
```

```python
df2['Total_Minutes'] = df2['Total_Minutes'].round(2)
```

```python
df2.head()
```

```python
df2.rename(columns={'Division':'Gender'}, inplace=True)
```

```python
df2.head()
```

```python
def get_lat_long(city, state):
  address = f"{city}, {state}"
  try:
        geolocator = Nominatim(user_agent="running", timeout=10)
        location = geolocator.geocode(address)
        if location:
            return location.latitude, location.longitude
        else:
            return None, None
  except GeocoderTimedOut:
        return None, None
```

```python
df2['Latitude'], df2['Longitude'] = zip(*df2.apply(lambda row: get_lat_long(row['City'], row['State']), axis=1))
```

```python
df2.head()
```

```python
df2['LatLong'] = df2['Latitude'].astype(str) + ', ' + df2['Longitude'].astype(str)
```

```python
df2.head()
```

```python
df2.to_csv('Cleaned_ff_race_50.csv', index=False)
```
